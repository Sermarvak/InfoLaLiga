```python
from selenium import webdriver
from selenium.common.exceptions import WebDriverException

from bs4 import BeautifulSoup
import requests

import time
```

```python
def driverChrome(url):
    driver = webdriver.Chrome('C:/Users/USUARIO/Desktop/ChromeDriver/chromedriver.exe')
    try:
        driver.get(url)
    except WebDriverException:
        driver.get(url)
    return driver
```

```python
# Función que obtiene el enlace de cada equipo
def obtenerEnlacesEquipos():    
    url_principal = 'https://www.laliga.com/laliga-easports/clubes'
    driver = driverChrome(url_principal)
    time.sleep(2)
    response = requests.get(url_principal)
    soup = BeautifulSoup(response.text, 'lxml')
    contenido = soup.find("div", class_="styled__GridStyled-sc-skzs8h-0 fVbFTE")
    elementos = contenido.find_all("a", class_="link")
    driver.close() 
    enlaces_equipos = []
    for i in range(len(elementos)):
        enlace = elementos[i].get('href')
        enlaces_equipos.append(enlace)
    return enlaces_equipos
```

```python
# Función que obtiene el enlace de cada jugador
def obtenerEnlacesJugadores():
    enlaces_jugadores = []
    enlaces_equipos = obtenerEnlacesEquipos()
    for i in range(len(enlaces_equipos)):
        url_equipo = 'https://www.laliga.com' + enlaces_equipos[i] + '/plantilla'
        driver = driverChrome(url_equipo)
        time.sleep(2)
        response = requests.get(url_equipo)
        soup = BeautifulSoup(response.text, 'lxml')
        contenido = soup.find("div", class_="styled__SquadListContainer-sc-sx1q1t-0 hwjXzG")
        elementos = contenido.find_all("a", class_="link")
        driver.close()
        for i in range(len(elementos)):
            enlace_jugador = elementos[i].get('href')
            enlaces_jugadores.append(enlace_jugador)
    return enlaces_jugadores
```
